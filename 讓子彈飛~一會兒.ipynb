```python
import pygame
import sys
from pygame import mixer
screen_width,screen_height = 1200,800
rgb = (255,255,0)
ship_v = [1,0]
alien_v = [0,0]
bullet_v = [0,-1]
aliens = []
bullets = []
ship,shipbox = None,None
mixer.pre_init(22100,-16,1,64)
mixer.init()
firesound = mixer.Sound("C:/Users/USER/Desktop/夢想的價值/Python/自學/gun-blast_A_minor.wav")
explosound = mixer.Sound("C:/Users/USER/Desktop/夢想的價值/Python/自學/shout-03.wav")
def is_collided(b1,b2):
    x1 = max(b1.left, b2.left)
    x2 = min(b1.right,b2.right)
    y1 = max(b1.top, b2.top)
    y2 = min(b1.bottom,b2.bottom)
    if x1<=x2 and y1<=y2:
        return True
    else:
        return False
def manage_sprites():
    global bullets,aliens
    for b in bullets:
        bulletbox = b[1]
        for a in aliens:
            alienbox = a[1]
            if is_collided(b[1],a[1]):
                bullets.remove(b)
                aliens.remove(a)
                explosound.play()
                break
    for b in bullets:
        if b[1].top<0:
            bullets.remove(b)
    print(len(bullets))
def create_sprite(imgfile):
    sprite = pygame.image.load(imgfile)
    bbox = sprite.get_rect()
    return sprite,bbox
def create_aliens(rows,cols,imgfile):
    for r in range(rows):
        for c in range(cols):
            alien,bbox = create_sprite(imgfile)
            bbox.left = c * 120 + 10
            bbox.top = r*(bbox.height + 10)
            aliens.append([alien,bbox])
            
def update_sprite(screen,sprite,bbox,v):
    bbox.centerx +=v[0]
    bbox.centery +=v[1]
    if bbox.left < 0:
        v[0] *= -1
    elif bbox.right>screen_width:
        v[0] *= -1
    screen.blit(sprite,bbox)
    
def update_bullets(screen,bullets):
    global bullet_v
    for b in bullets:
        update_sprite(screen, b[0],b[1], bullet_v)

def update_aliens(screen, aliens):
    global alien_v
    for a in aliens:
        update_sprite(screen, a[0], a[1], alien_v)
        
def check_events():
    global ship_v
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            mixer.quit()
            pygame.quit()
            sys.exit()
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                ship_v[0] = -1
            elif event.key == pygame.K_RIGHT:
                ship_v[0] = 1
            elif event.key == pygame.K_SPACE:
                if (len(bullets)<3):
                    bullet, bulletbox = create_sprite("C:/Users/USER/Desktop/夢想的價值/Python/自學/bullet1.png")
                    bulletbox.bottom = shipbox.top
                    bulletbox.centerx = shipbox.centerx
                    bullets.append([bullet, bulletbox])
                    firesound.play()
        elif event.type == pygame.KEYUP and event.key != pygame.K_SPACE:
            ship_v[0] = 0

def run_game(w,h,color):
    global ship,shipbox
    pygame.init()
    screen = pygame.display.set_mode((w,h))
    pygame.display.set_caption("MaryMin2")
    screen.fill(color)
    ship,shipbox = create_sprite("C:/Users/USER/Desktop/夢想的價值/Python/自學/帥.png")
    create_aliens(1,10,"C:/Users/USER/Desktop/夢想的價值/Python/自學/Balloon.png")
    shipbox.bottom = screen_height
    shipbox.centerx = screen_width//2
    screen.blit(ship,shipbox)
    while True:
        screen.fill(rgb)
        check_events()
        update_sprite(screen,ship,shipbox,ship_v)
        update_bullets(screen,bullets)
        update_aliens(screen,aliens)
        manage_sprites()
        pygame.display.flip()
#     pygame.quit()
    
run_game(screen_width,screen_height,rgb)


        
        






```
